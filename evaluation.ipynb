```python
import matplotlib.pyplot as plt
import numpy as np

DROPOUT = [0, 0.3, 0.6, 0.9]
ORI_LINEAR_ORI_BASIC = [0.607,0.597,0.544,0.5]
ORI_LINEAR_ORI_SUGGESTING_10 = [0.607,0.569,0.636,0.658]
ORI_LINEAR_ORI_SUGGESTING_20 = [0.607,0.575,0.575,0.606]

ORI_LINEAR_UCSD_BACIC = [0.773,0.719,0.564,0.5]
ORI_LINEAR_UCSD_SUGGESTING_10 = [0.773,0.740,0.766,0.838]
ORI_LINEAR_UCSD_SUGGESTING_20 = [0.773,0.739,0.626,0.749]

# Set the width of the bars
bar_width = 0.2

# Set the positions of the bars on the x-axis
index = np.arange(len(DROPOUT))

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Create bar plots for training on ORI and predicting ORI
bar1 = ax1.bar(index, ORI_LINEAR_ORI_BASIC, bar_width, label='BASIC ORI-UCSD', color='black')
bar2 = ax1.bar(index + bar_width, ORI_LINEAR_ORI_SUGGESTING_10, bar_width, label='OBJECTIVE ORI-UCSD 10', color='navy')
bar3 = ax1.bar(index + 2 * bar_width, ORI_LINEAR_ORI_SUGGESTING_20, bar_width, label='OBJECTIVE ORI-UCSD 20', color='grey')

# Customize the first subplot
ax1.set_xlabel('Dropout')
ax1.set_ylabel('AUC')
ax1.set_title('Training on ORI and Predicting ORI')
ax1.set_xticks(index + bar_width)
ax1.set_xticklabels(DROPOUT)

# Create bar plots for training on ORI and predicting UCSD
bar4 = ax2.bar(index, ORI_LINEAR_UCSD_BACIC, bar_width, color='black')
bar5 = ax2.bar(index + bar_width, ORI_LINEAR_UCSD_SUGGESTING_10, bar_width, color='navy')
bar6 = ax2.bar(index + 2 * bar_width, ORI_LINEAR_UCSD_SUGGESTING_20, bar_width, color='grey')

# Customize the second subplot
ax2.set_xlabel('Dropout')
ax2.set_ylabel('AUC')
ax2.set_title('Training on ORI and Predicting UCSD')
ax2.set_xticks(index + bar_width)
ax2.set_xticklabels(DROPOUT)

# Place a single label on the side
fig.text(0.98, 0.5, 'AUC', va='center', rotation='vertical')

# Place the legend outside the plot
fig.legend([bar1, bar2, bar3], ['BASIC', 'OBJECTIVE 10', 'OBJECTIVE 20'], loc='upper left', bbox_to_anchor=(1.05, 1))

plt.tight_layout()
plt.show()
```
